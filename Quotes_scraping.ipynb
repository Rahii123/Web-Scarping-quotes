```python
import requests
from bs4 import BeautifulSoup

url = "https://quotes.toscrape.com/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

quotes = soup.find_all("div", class_="quote")

print("Number of quotes found:", len(quotes))
print("Example of one quote tag:", quotes[0])
```

```python
import pandas as pd

# Create a list to hold all data
data = []

# Loop through each quote block
for quote in quotes:
    text = quote.find("span", class_="text").text.strip()
    author = quote.find("small", class_="author").text.strip()
    tags = [tag.text for tag in quote.find_all("a", class_="tag")]
    tags_str = ", ".join(tags)

    data.append({
        "Quote": text,
        "Author": author,
        "Tags": tags_str
    })

# Create a DataFrame and save to CSV
df = pd.DataFrame(data)
df.to_csv("quotes.csv", index=False)

print("✅ Data saved to quotes.csv")
df.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

all_data = []
page = 1

while True:
    url = f"https://quotes.toscrape.com/page/{page}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    quotes = soup.find_all("div", class_="quote")
    if not quotes:
        break  # No quotes means we reached last page

    for quote in quotes:
        text = quote.find("span", class_="text").text.strip()
        author = quote.find("small", class_="author").text.strip()
        tags = [tag.text for tag in quote.find_all("a", class_="tag")]
        tags_str = ", ".join(tags)
        all_data.append({
            "Quote": text,
            "Author": author,
            "Tags": tags_str
        })

    page += 1

df = pd.DataFrame(all_data)
df.to_csv("all_quotes.csv", index=False)
print("✅ All quotes scraped and saved to all_quotes.csv")
df.head(25)
```
